```python
# read it in to inspect it
with open('input.txt', 'r', encoding='utf-8') as f:
    text = f.read()
```

```python
print("length of dataset in characters: ", len(text))
```

```python
print(text[:1000])
```

```python
# here are all the unique characters that occur in this text
chars = sorted(list(set(text)))
vocab_size = len(chars)
print(''.join(chars))
print(vocab_size)
```

```python
# create a mapping from characters to integers
stoi = { ch:i for i,ch in enumerate(chars) }
print(stoi)
itos = { i:ch for i, ch in enumerate(chars) }
print(itos)
encode = lambda s: [stoi[c] for c in s] # encoder: take a string, output a list of integers
# def encode(s):
#    result = []
#    for c in s:
#        result.append(stoi[c])
#    return result

decode = lambda l: ''.join([itos[i] for i in l]) # decoder: take a list of integers, output a string
# def decode(l):
#    result = []
#    for i in l:
#        result.append(itos[i])
#    return result

print(encode("hi there"))
print(decode(encode("hi there")))
```

```python
# let's now encode the entire text dataset and store it into a torch.tensor
import torch
data = torch.tensor(encode(text), dtype=torch.long)
print(data.shape, data.dtype)
print(data[:1000])
```

```python
# Let's now split up the data into train and validation sets
n = int(0.9 * len(data)) # first 90% will be train, rest val
train_data = data[:n]
val_data = data[n:]
```

```python
block_size = 8 # 每个样本的长度
train_data[:block_size+1]
```

```python
x = train_data[:block_size] # 输入的字符
y = train_data[1:block_size + 1] # 预测的字符
for t in range(block_size):
  context = x[:t + 1]
  target = y[t]
  print(f"when input is {context} the target: {target}")
```

```python
torch.manual_seed(1337)
batch_size = 4 # 一次取4个样本
block_size = 8 # 每个样本的长度

def get_batch(split):
    data = train_data if split == 'train' else val_data
    ix = torch.randint(len(data) - block_size, (batch_size,)) # 起点索引
    x = torch.stack([data[i:i+block_size] for i in ix])
    y = torch.stack([data[i + 1 : i + block_size + 1] for i in ix])
    # 从data中取block_size个字符（列数），堆叠batch_size个样本（行数）
    return x, y

xb, yb = get_batch('train')
print('inputs: ')
print(xb.shape)
print(xb)
print('targets: ')
print(yb.shape)
print(yb)

print('-----')

for b in range(batch_size): # 挑选行
    for t in range(block_size): # 列范围
        context = xb[b, :t+1]
        target = yb[b, t]
        print(f"when input is {context.tolist()} the target: {target}")
```

```python
import torch
import torch.nn as nn
from torch.nn import functional as F
torch.manual_seed(1337)

class BigramLanguageModel(nn.Module):
    def __init__(self, vocab_size): # 初始化方法
        super().__init__() # 调用父类的初始化方法
        self.token_embedding_table = nn.Embedding(vocab_size, vocab_size) # 随机初始化矩阵（权重）
        
    def forward(self, idx, targets): # 前向传播方法
        logits = self.token_embedding_table(idx) # 查表得到词向量

        B, T, C = logits.shape  
        logits = logits.view(B*T, C) # 展平张量，每个位置的预测概率分布
        targets = targets.view(B*T) # 展平张量，每个位置的目标值
        loss = F.cross_entropy(logits, targets) # 计算交叉熵损失

        return logits, loss

    def generate(self, idx, max_new_tokens):
        for _ in range(max_new_tokens):
            logits, loss = self(idx)

m = BigramLanguageModel(vocab_size)
logits, loss =  m(xb, yb)
print(logits.shape)
print(loss)
```
